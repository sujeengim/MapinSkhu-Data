kwan 모델을 위한 데이터 만들기

- kwan_num
- kwan_name

x kwan_image

```python
import pandas as pd

import numpy as np
```

```python
csv = pd.read_csv('room.csv')
```

```python
data = csv.drop(csv.columns[0], axis=1)
```

```python
data
```

```python
df = data[['kwan_name']]
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
## kwan_num 생성

conditionlist = [
    (df['kwan_name'] == '승연관'),
    (df['kwan_name'] == '월당관'), 
    (df['kwan_name'] == '미가엘관'),
    (df['kwan_name'] == '성미가엘성당&피츠버그홀'),
    (df['kwan_name'] == '새천년관'),
    (df['kwan_name'] == '정보과학관'),
    (df['kwan_name'] == '일만관'),
    (df['kwan_name'] == '나눔관')
]
choicelist = [
    '1',
    '3',
    'M',
    '9',
    '7',
    '6',
    'B',
    '5C'
]

df['kwan_num'] = np.select(conditionlist, choicelist)
```

```python
df
```

```python
data = df[['kwan_num', 'kwan_name']]
```

```python
data
```

```python
# kwan 데이터 저장

data.to_csv('kwan.csv', encoding='utf-8-sig', index=None)
```
