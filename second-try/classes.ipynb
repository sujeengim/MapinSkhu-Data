```python
import pandas as pd

import numpy as np
```

```python
csv = pd.read_csv('data2.csv')
```

```python
## 데이터 준비

data = csv.drop([csv.columns[0]], axis=1)
```

```python
data
```

```python
df = data.copy(deep=True)
```

```python
## kwan_name 컬럼 생성

df['kwan_name'] = df['room'].str[0]

conditionlist = [
    (df['kwan_name'] == '1'),
    (df['kwan_name'] == '3'), 
    (df['kwan_name'] == 'M'),
    (df['kwan_name'] == '9'),
    (df['kwan_name'] == '7'),
    (df['kwan_name'] == '6'),
    (df['kwan_name'] == 'B'),
    (df['kwan_name'] == '5')
] #첫 문자로 관 판별
choicelist = [
    '승연관',
    '월당관',
    '미가엘관',
    '성미가엘성당&피츠버그홀',
    '새천년관',
    '정보과학관',
    '일만관',
    '나눔관'
]

df['kwan'] = np.select(conditionlist, choicelist, default='미정') # room이 미정인거 처리
```

```python
df['kwan_name'] = df['kwan']
```

```python
data = df.drop([df.columns[-1]], axis=1)
```
