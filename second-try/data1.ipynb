```python
import pdfplumber as pp

import pandas as pd
```

```python
pdf = pp.open("23-1.pdf") 
pdf #pdf 한페이지씩 돌면서 각 페이지 테이블 추출해야함
```

```python
data = [] #테이블 저장할 리스트
```

```python
page_len = len(pdf.pages) #페이지 수 구하기 

for p in range(page_len):
    table_len = len(pdf.pages[p].extract_tables()) #한 페이지 내에서 테이블개수 출력
    
    for t in range(table_len): 
        pre = pdf.pages[p].extract_tables() 
        data.append(pd.DataFrame(pre[t])) #테이블 하나씩 데이터프레임으로 변환후 data에 담기
```

```python
## 헤더 지정해주기

for i in range(len(data)):
    df = data[i]
    df_column = df.iloc[0]
    df = df[1:]
    df.columns = df_column
    data[i] = df
```

```python
data
```

```python
## 변수명 변경

for i in range(len(data)):
    df = data[i]
    
    check_list = ['이수\n구분', '과목코드', '교과목명', '교수명', '강의실/시간']
    
    for j in check_list:
        if j not in df.columns:
            print(i, '는', j, '가 없다.' )
```

```python
for i in range(len(data)):
    data[i] = data[i].rename(columns={
        '이수\n구분':'kind',
        '과목코드':'code',
        '교과목명':'class_name',
        '교수명':'prof',
        '강의실/시간':'room_time'
    })
```

```python
## 필요한 변수만 남기기

for i in range(len(data)):
    df = data[i]
    df = df[['kind', 'code', 'class_name', 'prof', 'room_time']]
    data[i] = df
```
