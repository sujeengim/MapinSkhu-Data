```python
## 작업 전 먼저 공백 제거
# data['room_time'].str.replace(" ", "", )

for col in ['kind', 'code', 'class_name', 'prof', 'room_time']:
    df = data[col].copy(deep=True)
    data[col] = df.str.replace(" ", "")
```

```python
data
```

```python
## 개행문자 기준으로 변수 나누기

data['room_time2'] = data.room_time.str.split('\n').str[1]
data['room_time'] = data.room_time.str.split('\n').str[0]
```

```python
data
```

```python
data.info()
```

```python
## date1&date2 time 파생변수 생성

df = data.copy(deep=True)

df['time1'] = df.room_time.str.split('/').str[2]
df['time2'] = df.room_time2.str.split('/').str[2]
```

```python
df.tail()
```

```python
## time1&time2가 nan이 아닌데 time1!=time2인 데이터는 nan으로 result에 저장

df['result'] = np.where(
    (pd.notna((df['time2'])) & (df['time1'] != df['time2']) & (pd.notna(df['time1']))),
    np.nan,
    df['time1']
)
```

```python
df.tail()
```

```python
# time1 != time2 데이터 추출
df[df.result.isna()]
```

```python
## date1&date2 room 파생변수 생성

df = data.copy(deep=True)

df['room'] = df.room_time.str.split('/').str[0]
df['room2'] = df.room_time2.str.split('/').str[0]
```

```python
df.tail()
```

```python
## room&room2가 nan이 아닌데 room!=room2인 데이터는 nan으로 result에 저장

df['result'] = np.where(
    (pd.notna((df['room'])) & (df['room'] != df['room2']) & (pd.notna(df['room2']))),
    np.nan,
    df['room']
)
```

```python
# room != room2 데이터 추출
df[df.result.isna()]
```

```python
data['room'] = df['result']
```

```python
## 강의실 미정, 이상치 데이터 추출

df = data.copy(deep=True)
df[df['room'].str.contains('9999') | df['room'].str.contains('미정')]
```

```python
#kind 제외하고 중복 확인
df['dup2'] = df.duplicated(['code', 'class_name', 'prof', 'room', 'date1', 'date2', 'start', 'end'], keep=False)
```

# 여기부터 바뀜

```python
dup_df_sort = df[df['dup2']==True].sort_values('code') #코드기준 정렬
```

```python
len(dup_df_sort)
```

```python
dup_dfs = [] #처리 잘됐는지 확인 위해 생성함
x = 0

# 두개씩 겹치는 수업밖에 없음. 겹치는 수업 두개씩 묶어 저장한다.
for _ in range(len(dup_df_sort)//2):
    test = dup_df_sort.iloc[x:x+2]
    dup_dfs.append(test)
    x += 2
```

```python
dup_dfs[-1]
```

```python
# 처리한 값 저장하기
dup_list = []

for test in dup_dfs:
    s1, s2 = test['kind'] #kind값만 추출해서
    value = ', '.join([s1, s2]) #합쳐버리기
    test['kind'] = value
    
    s_index = test.index.to_list()
    df.loc[s_index] = test
    
    dup_list.append(test)
```
