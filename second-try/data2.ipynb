```python
# "room/data/start:end"형식 안 맞는 행 찾기
series = data['room_time']

for i in range(len(series)):
    cnt = 0
    for j in series[i]:
        if j=='/':
            cnt+=1
    if cnt != 2 and cnt != 4:
        print(series[[i]])
```

```python
# room_time 형식 맞추기  - 한번에 하는 방법..생각해보기
data.at[12, 'room_time'] = '7207/수/10:30~11:45'
data.at[13, 'room_time'] = '7207/목/10:30~11:45'
data.at[168, 'room_time'] = '7208/목/09:00~11:50'
data.at[229, 'room_time'] = '3402/수/15:00~17:50'
data.at[230, 'room_time'] = '3301/목/12:00~14:50'
```

```python
# 확인
data.loc[[12,13,168,229,230]]
```

```python
## 작업 전 먼저 공백 제거
# data['room_time'].str.replace(" ", "", )

for col in ['kind', 'code', 'class_name', 'prof', 'room_time']:
    df = data[col].copy(deep=True)
    data[col] = df.str.replace(" ", "")
```

```python
data
```

```python
## 개행문자 기준으로 변수 나누기

data['room_time2'] = data.room_time.str.split('\n').str[1]
data['room_time'] = data.room_time.str.split('\n').str[0]
```

```python
data
```

```python
data.info()
```

```python
## date1&date2 time 파생변수 생성

df = data.copy(deep=True)

df['time1'] = df.room_time.str.split('/').str[2]
df['time2'] = df.room_time2.str.split('/').str[2]
```

```python
df.tail()
```

```python
## time1&time2가 nan이 아닌데 time1!=time2인 데이터는 nan으로 result에 저장

df['result'] = np.where(
    (pd.notna((df['time2'])) & (df['time1'] != df['time2']) & (pd.notna(df['time1']))),
    np.nan,
    df['time1']
)
```

```python
df.tail()
```

```python
# time1 != time2 데이터 추출
df[df.result.isna()]
```

```python
## date1&date2 room 파생변수 생성

df = data.copy(deep=True)

df['room'] = df.room_time.str.split('/').str[0]
df['room2'] = df.room_time2.str.split('/').str[0]
```

```python
df.tail()
```

```python
## room&room2가 nan이 아닌데 room!=room2인 데이터는 nan으로 result에 저장

df['result'] = np.where(
    (pd.notna((df['room'])) & (df['room'] != df['room2']) & (pd.notna(df['room2']))),
    np.nan,
    df['room']
)
```

```python
# room != room2 데이터 추출
df[df.result.isna()]
```

```python
data['room'] = df['result']
```

```python
## 강의실 미정, 이상치 데이터 추출

df = data.copy(deep=True)
df[df['room'].str.contains('9999') | df['room'].str.contains('미정')]
```

```python
## date1, date2 파생변수 생성

df = data.copy(deep=True)

data['date1'] = df.room_time.str.split('/').str[1]
data['date2'] = df.room_time2.str.split('/').str[1]
```

```python
## time1, time2 파생변수 생성
```

```python
df = data.copy(deep=True)

df['time'] = df.room_time.str.split('/').str[2]
data['start'] = df.time.str.split('~').str[0]
data['end'] = df.time.str.split('~').str[1]
```

```python
data.tail()
```

```python
## 필요한 컬럼만 남기기

data = data[['kind', 'code', 'class_name', 'prof', 'room', 'date1', 'date2', 'start', 'end']]
data.tail()
```

```python
## 중복데이터 제거 
```

```python
df = data.copy(deep=True)

# 중복데이터 찾기
df['dup'] = df.duplicated(['kind', 'code', 'class_name', 'prof', 'room', 'date1', 'date2', 'start', 'end'], keep=False)
df[df['dup']==True].sort_values('code')
```

```python
# 중복 제거
df = df.drop_duplicates()
```
