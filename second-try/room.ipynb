room 모델을 위한 데이터 만들기

- kwan_name
- room
- floor
- room_image
- room_type ()

```python
import pandas as pd
import numpy as np
```

```python
csv = pd.read_csv('classes.csv')
```

```python
csv
```

```python
data = csv.drop(csv.columns[0], axis=1)
```

```python
data
```

```python
df = data[['room', 'kwan_name']]
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df['floor'] = np.where(
    df['room'] == '미정',
    None,
    df['room'].str[1]
) 
```

```python
data
```

```python
df = df.dropna(axis=0)
```

```python
df
```

```python
# 두 번째 자리 문자가 숫자가 아닌 경우는 c밖에 없음
df.loc[df['floor']=='C', 'floor'] = df['room'].str[2]
```

```python
df
```

```python
len(df)
```

```python
df = df.drop_duplicates()
```

```python
len(df)
```

```python
df['room_image'] = None
```

```python
df['room_type'] = None
```

```python
df
```

```python
data = df[['kwan_name','room','floor','room_image','room_type']]
```

```python
data
```

```python
# room 데이터 저장
```

```python
data.to_csv('room.csv', encoding='utf-8-sig', index=None)
```
