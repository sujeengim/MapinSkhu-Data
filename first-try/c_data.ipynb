```python
import pdfplumber as pp

pdf = pp.open("23-1_책자.pdf") #pdf 불러오기
pdf
```

```python
## 데이터 이상 생겼을 때 여기서부터 다시 실행 

import pandas as pd

data = [] #테이블 저장할 리스트

page_len = len(pdf.pages)

for p in range(page_len):
    if p==34:
        continue
    # len(pdf.pages[p].extract_table()) #행개수 출력
    table_len = len(pdf.pages[p].extract_tables()) #테이블개수 출력
    
    for t in range(table_len):
        pre = pdf.pages[p].extract_tables() #페이지 내 모든 테이블 저장
        pre_df = pd.DataFrame(pre[t]) #테이블 하나씩 가져오기
        
        pre_df2 = pre_df.rename(columns=pre_df.iloc[0]) #첫 행을 컬럼으로 설정
        df = pre_df2.drop(pre_df2.index[0]) # 첫 행 삭제
        
        data.append(df) #데이터 리스트에 추가
        
data
```

```python
## 필요한 속성만 남기기

for i in range(len(data)):
    df = data[i]
    df.rename(columns={'이수\n구분':'이수구분'}, inplace=True)
    df = df[['이수구분', '과목코드', '교과목명', '교수명', '강의실/시간']]
    
    data[i] = df
    
data[0]
```

```python
## 모든 테이블 하나로 합치기 


data = pd.concat(data, ignore_index=True) #ignore_index=True을 줘서 인덱스를 재배열
data
```

```python
## 변수명 변경

data = data.rename(columns={
    "이수구분":"kind",
    "과목코드":"code",
    "교과목명":"class_name",
    "교수명":"prof_name",
    "강의실/시간":"room_time"
})
data
```

```python
import numpy as np
```

```python
data['room_time'] = data['room_time'].str.replace(" ", "", ) #공백문자 제거. 세 번째 매개변수 비워두면 모든 공백 제거
# data = data['room_time2'].str.replace(" ", "", )
data
```

```python
## 문자열 분할후 다른 변수에 저장하기. 순서중요

data['room_time2'] = data.room_time.str.split('\n').str[1]
data['room_time'] = data.room_time.str.split('\n').str[0]
data
```

```python
data['room'] = data.room_time.str.split('/').str[0]

data['date1'] = data.room_time.str.split('/').str[1]
data['date2'] = data.room_time2.str.split('/').str[1]

data['time'] = data.room_time.str.split('/').str[2]
data['start'] = data.time.str.split('~').str[0]
data['end'] = data.time.str.split('~').str[1]

# 필요한 속성만 남기기
data = data[['kind', 'code', 'class_name', 'prof_name', 'room', 'date1', 'date2', 'start', 'end']]
data
```

```python
## 데이터 저장해두기

data.to_csv("data.csv", encoding='utf-8-sig') #index포함해서 저장할거니까 index=True(기본값)
```
